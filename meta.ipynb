```python
from collections import Counter
```

```python
genres = dict(Counter(df['Genre']))
genres = pd.DataFrame.from_dict(genres, orient='index')
genres.sort_values(by = 0,ascending=False)
```

# Genres

```python
genres.plot(kind = 'bar',subplots=True,figsize = (16,10),fontsize=28)
```

# Singers' genders

```python
insts = dict(Counter(df['Vocal']))
insts = pd.DataFrame.from_dict(insts,orient = 'index')
insts.plot(kind ='bar',subplots=True)
```
